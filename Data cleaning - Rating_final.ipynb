```python
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
off_rating = pd.read_csv("Office_rating.csv") 
off_rating
```

```python
off_meta = pd.read_csv("Office_meta.csv") 
off_meta
```

```python
off_merged = off_rating.merge(off_meta, how = 'left', on = 'asin')
off_merged
```

```python
off_merged.isnull().sum()
```

```python
off_merged = off_merged.dropna(subset = 'main_cat')
off_merged.isnull().sum()
```

```python
off_merged['title'] = off_merged['title'].fillna("Unknown")
off_merged['brand'] = off_merged['brand'].fillna(off_merged['brand'].mode()[0])
off_merged.isnull().sum()
```

```python
cds_rating = pd.read_csv("CdsVinyl_rating.csv") 
cds_rating
```

```python
cds_rating.isnull().sum()
```

```python
cds_meta = pd.read_csv("CdsVinyl_meta.csv") 
cds_meta
```

```python
cds_merged = cds_rating.merge(cds_meta, how = 'left', on = 'asin')
cds_merged
```

```python
cds_merged.isnull().sum()
```

```python
cds_merged = cds_merged.dropna(subset = "main_cat")
cds_merged
```

```python
cds_merged.isnull().sum()
```

```python
cds_merged['title'] = cds_merged['title'].fillna("Unknown")
cds_merged['brand'] = cds_merged['brand'].fillna(cds_merged['brand'].mode()[0])
cds_merged.isnull().sum()
```

```python
data = [off_merged, cds_merged]
rating = pd.concat(data,ignore_index = True)
rating
```

```python
rating.isnull().sum()
```

```python
data = rating[['rating','price','rank']]
```

```python
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
data = data
imp_mean = IterativeImputer(random_state=0)
imp_mean.fit(data)
IterativeImputer(random_state = 10)
a = imp_mean.transform(data)
a
```

```python
a = pd.DataFrame(a)
a.rename(columns = {0:'overall', 1:"price", 2:"rank"}, inplace = True)
a
```

```python
rating['price'] = a.price
rating['rank'] = a['rank']
```
