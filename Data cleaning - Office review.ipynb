```python
from datetime import datetime
import re 
import numpy as np 
import pandas as pd
import gzip
import json 
```

```python
# User defined function to extract the zipped Json file and form a Dataframe 

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
def categorise(row):  
    
    if row['overall'] == 1 :
        return 'poor'
    elif row['overall'] == 2 :
        return 'bad'
    elif row['overall'] == 3 :
        return 'average'
    elif row['overall'] == 4 :
        return 'good'
    elif row['overall'] == 5 :
        return 'great'
    
off['class'] = off.apply(categorise, axis=1)
off.head()
```

```python
off.isnull().sum()
```

```python
off.reviewText = off.reviewText.fillna(categorise)
off.isnull().sum()
```

```python
def cat_summary(row):  
    
    if row['overall'] == 1 :
        return 'One Star'
    elif row['overall'] == 2 :
        return 'Two Stars'
    elif row['overall'] == 3 :
        return 'Three Stars'
    elif row['overall'] == 4 :
        return 'Four Stars'
    elif row['overall'] == 5 :
        return 'Five Stars'
```

```python
off.summary = off.summary.fillna(cat_summary)
```
