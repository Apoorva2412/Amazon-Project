```python
from datetime import datetime
import re 
import numpy as np
import pandas as pd
import gzip
import json 

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

# Cleaning Price column

```python
cds_meta['price'].value_counts()
```

```python
price = []
for i in cds_meta['price']:
    if len(i) < 10:
        price.append(i)
    else:
        i = 0
        price.append(i)
#price        
cds_meta['price'] = pd.DataFrame(price)
cds_meta['price']
```

```python
cds_meta['price'] = pd.to_numeric(cds_meta['price'], downcast="float")
```
