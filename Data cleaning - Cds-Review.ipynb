```python
from datetime import datetime
import re 
import numpy as np
```

```python
import pandas as pd
import gzip
import json 

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield json.loads(l)

def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
cds_data = getDF("A:/NIIT/Project - Amazon Reviews/CDs_and_Vinyl_5.json.gz")
cds_data
```

```python
cds = cds_data.copy()
```

```python
cds['reviewTime'] = pd.to_datetime(cds['reviewTime'])
cds.drop(["vote","image","unixReviewTime",'style'], axis = 1, inplace = True)
```

```python
cds.duplicated().sum()
```

```python
cds = cds.drop_duplicates(keep = 'first', ignore_index = True)
cds
```

```python
cds.isnull().sum()
```

```python
cds['reviewText'] = cds['reviewText'].str.lower()
cds['reviewText'].value_counts()
```

```python
def categorise(row):  
    
    if row['overall'] == 1 :
        return 'poor'
    elif row['overall'] == 2 :
        return 'bad'
    elif row['overall'] == 3 :
        return 'average'
    elif row['overall'] == 4 :
        return 'good'
    elif row['overall'] == 5 :
        return 'great'
    
cds['class'] = cds.apply(categorise, axis=1)
```

```python
cds.reviewText.value_counts()
```

```python
cds.reviewText = cds.reviewText.fillna(categorise)
```

```python
cds.isnull().sum()
```

```python
def cat_summary(row):  
    
    if row['overall'] == 1 :
        return 'One Star'
    elif row['overall'] == 2 :
        return 'Two Stars'
    elif row['overall'] == 3 :
        return 'Three Stars'
    elif row['overall'] == 4 :
        return 'Four Stars'
    elif row['overall'] == 5 :
        return 'Five Stars'
```

```python
cds.summary = cds.summary.fillna(cat_summary)
```
