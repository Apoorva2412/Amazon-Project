```python
import pandas as pd
from datetime import datetime
```

```python
cds_rating = pd.read_csv("A:/NIIT/Project - Amazon Reviews/CDs_and_Vinyl.csv", header = None)
cds_rating.head()
```

```python
cds_rating.rename(columns = {0:"asin", 1:"reviewerID", 2:"rating", 3:"date"}, inplace = True)
cds_rating.head()
```

```python
date = []
for i in cds_rating['date']:
    i = datetime.utcfromtimestamp(i).strftime('%Y-%m-%d')
    date.append(i)
date
```

```python
cds_rating.date = date
cds_rating.head()
```

```python
cds_rating.isnull().sum()
```

```python
cds_rating.drop_duplicates(keep='first', inplace = True)
cds_rating
```
