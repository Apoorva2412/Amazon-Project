```python
import pandas as pd
```

```python
off_review = pd.read_csv("Office_reviews.csv")
off_review
```

```python
off_review.isnull().sum()
```

```python
off_review.reviewerName = off_review.reviewerName.fillna("Amazon Customer")
off_review.isnull().sum()
```

```python
def cat_summary(row):  
    
    if row['overall'] == 1 :
        return 'One Star'
    elif row['overall'] == 2 :
        return 'Two Stars'
    elif row['overall'] == 3 :
        return 'Three Stars'
    elif row['overall'] == 4 :
        return 'Four Stars'
    elif row['overall'] == 5 :
        return 'Five Stars'
```

```python
off_review.summary = off_review.summary.fillna(cat_summary)
off_review.isnull().sum()
```

```python
off_meta = pd.read_csv("Office_meta.csv")
off_meta
```

```python
off_meta.isnull().sum()
```

```python
off_meta['brand'] = off_meta['brand'].fillna(off_meta['brand'].mode()[0])
off_meta['title'] = off_meta['title'].fillna("unknown")
#off_meta['rank'] = off_meta['rank'].fillna(0)

off_meta.isnull().sum()
```

```python
# off_merged = pd.merge(off_review , off_meta, how = 'left', on = 'asin')
# off_merged
off_merged = off_review.merge(off_meta, left_on = 'asin', right_on = 'asin')
off_merged
```

```python
cds_review = pd.read_csv("CdsVinyl_reviews.csv") 
cds_review
```

```python
cds_review.isnull().sum()
```

```python
cds_review.reviewerName = cds_review.reviewerName.fillna("Amazon Customer")
cds_review.summary = cds_review.summary.fillna(cat_summary)
cds_review.isnull().sum()
```

```python
def categorise(row):  
    
    if row['overall'] == 1 :
        return 'poor'
    elif row['overall'] == 2 :
        return 'bad'
    elif row['overall'] == 3 :
        return 'average'
    elif row['overall'] == 4 :
        return 'good'
    elif row['overall'] == 5 :
        return 'great'
    
```

```python
cds_review.reviewText = cds_review.reviewText.fillna(categorise)
cds_review.isnull().sum()
```

```python
cds_meta = pd.read_csv("CdsVinyl_meta.csv") 
cds_meta
```

```python
cds_meta.isnull().sum()
```
