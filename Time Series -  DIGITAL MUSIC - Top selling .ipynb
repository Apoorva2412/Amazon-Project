# Checking for data stationarity as stationary data

```python
def checkStationarity(data):
    pvalue = adfuller(data)[1]
    print(pvalue)
    if pvalue < 0.05:
        ret = "Data is Stationary. Proceed to model building"
    else:
        ret = "Data is not Stationary. Make is stationary"
        
    return(ret)
```

```python
checkStationarity(office_df)
```

```python
diff_office_df = office_df - office_df.shift(1)
diff_office_df.dropna(inplace = True)
checkStationarity(diff_office_df)
```

# Train-Test Split data

```python
# split the data into train and test
length = int(0.90 * len(office_df))
train2 = office_df.iloc[:length]
test2 = office_df.iloc[length:]
print("train = {}, test = {}".format(len(train2), len(test2)))
```

# Hyper-parameter tunning for Modelling

```python
d = 1; 
lis=[]
for p in range(3):
    for q in range(2):
        lis.append((p,d,q))

# building model
model=[]
for i in lis:
    m = ARIMA(train2, order=i).fit() 
    model.append(m)

# checking model is good or not
c=1
for i in model:
    pvalue = sm.stats.acorr_ljungbox(i.resid, lags=[1], return_df=True)['lb_pvalue'].values
    if pvalue < 0.05:
        print(f'Model {c} Not a good model')
    else:
        print(f'Model {c} Good model')
    c+=1
```

```python
# calculating AIC and BIC score   
aic=[]
bic=[]
for i in model[0:]:
    aic.append(i.aic)
    bic.append(i.bic)

# predicting values
forecast=[]
for i in model[0:]:
    fcast2 = i.predict(start=len(train2), end=len(train2)+len(test2)-1)
    forecast.append(fcast2)

# calculating MSE and RMSE score
MSE = []
RMSE = []
for i in forecast:
    mse = mean_squared_error(test2, i)
    rmse = np.sqrt(mse)
    MSE.append(mse)
    RMSE.append(rmse)

# printing model numbers
model_no=[]
for i in range(0,len(model)):
    model_no.append('Model '+str(i))
    
# dataframe of all results
df_results = pd.DataFrame({'Model No.': model_no, '(p, d, q)':lis[0:], 'AIC': aic, 'BIC': bic, 'MSE': MSE, 'RMSE': RMSE})
df_results
```

```python
print('Taking model with lowest RMSE score:')
df_results[df_results.RMSE==df_results.RMSE.min()]
```

# SARIMA model

```python
p=0 ; q=0; d=1
P=0; Q=0
m2 = sm.tsa.statespace.SARIMAX(train2,order=(p,d,q),seasonal_order=(P,d,Q,12)).fit()
```

```python
# predict

predictions = m2.predict(start,end,dynamic=True)
df_res = pd.DataFrame({'actual':test2, 'predictions':predictions})
df_res['err'] = df_res.actual - df_res.predictions
#print(df_res)
print(df_res)
# MSE
mse4 = mean_squared_error(df_res.actual,df_res.predictions)
print('mse=',mse4)
rmse4 = np.sqrt(mse4)
print('root mean squared error =',rmse4)
```
